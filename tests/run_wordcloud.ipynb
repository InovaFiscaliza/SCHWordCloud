```python
credentials = config['api_credentials']['google_search']['maxwelfreitas']
gs = schwordcloud.GoogleSearch(credentials)
gs
```

```python
for item in dm.items_to_search[:2]:
    result = gs.request_wordcloud(item)
    result_status_code = result['status_code']
    if result_status_code == 200:
        # add result and format annotation
        dm.add_result(result)
        dm.items_to_search.remove(item)
```
